3. Get a random sample of size 60 ( This should contains 20 randon data points from each species type). Write a simple function to calculate the mean and he standard deviation of SW, SL, PW and PL from that sample for each species types.

```python
# Create a dictionary to store the means and standard deviations for each species
result = {}
```

```python
# Define a function to calculate mean and standard deviation for a given feature
def calculate_mean_and_std(data, feature):
    mean = data[feature].mean()
    std = data[feature].std()
    return mean, std
```

```python
import random
```

```python
# Create a random sample of 20 data points for each species
for species in df['Iris type'].unique():
    species_data = df[df['Iris type'] == species]
    random_indices = random.sample(range(len(species_data)), 20)
    random_sample = species_data.iloc[random_indices]

    # Calculate mean and standard deviation for each feature
    means_std = {}
    for feature in ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']:
        mean, std = calculate_mean_and_std(random_sample, feature)
        means_std[feature] = {'mean': mean, 'std': std}

    result[species] = means_std
```

```python
# Print the results
for species, values in result.items():
    print(f"Species: {species}")
    for feature, stats in values.items():
        print(f"{feature} - Mean: {stats['mean']:.2f}, Std Dev: {stats['std']:.2f}")
    print()
```
